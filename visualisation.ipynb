```python
#foncton pour reverser sumcum()
def anti_sumacc(l):
    l= l.tolist()
    l = np.array(l,dtype = int)
    i = 1
    l2 = np.zeros(len(l))
    while i < len(l):
        if l[i] != 0:
            l2[i] = abs(l[i] - l[i-1].sum())
        i = i+1
    l2[0]=l[0]
    return l2
print('fonction crée avec success')
```

```python
#appliquer la fonction anti_sumacc sur les donné de la region casa_settat
data_casa2 = anti_sumacc(data_casa)
data_casa2 = pd.DataFrame({'Casablanca-Settat':data_casa2})
data_casa2.head(5)
```

```python
# normalize Confirmed data
norm_Confirmed = (data_maroc_global_info['Confirmed']-data_maroc_global_info['Confirmed'].min())/(data_maroc_global_info['Confirmed'].max()-data_maroc_global_info['Confirmed'].min())

# normalize Deaths data
norm_Deaths =(data_maroc_global_info['Deaths']-data_maroc_global_info['Deaths'].min())/(data_maroc_global_info['Deaths'].max()-data_maroc_global_info['Deaths'].min())

# normalize Recovered  data
norm_Recovered =(data_maroc_global_info['Recovered']-data_maroc_global_info['Recovered'].min())/(data_maroc_global_info['Recovered'].max()-data_maroc_global_info['Recovered'].min())

d = {'Confirmed': norm_Confirmed, 'Deaths': norm_Deaths, 'Recovered':norm_Recovered,}
df_c_d = pd.DataFrame(data=d)

df_c_d.reset_index(inplace = True) 
df_c_d.columns = list(map(str, df_c_d.columns))
```

```python
#code plot scatter 
ax0 = df_c_d.plot(kind='scatter',x='Date',y='Confirmed',figsize=(14, 8),alpha=0.5,color='blue', s=norm_Confirmed * 2000 + 10)
df_c_d.plot(kind='scatter', x='Date',y='Deaths',alpha=0.5, color="red",s=norm_Deaths* 2000 + 10, ax = ax0)
df_c_d.plot(kind='scatter', x='Date',y='Recovered',alpha=0.5, color="green",s=norm_Recovered* 2000 + 10, ax = ax0)
ax0.set_ylabel('nombre des cas')
ax0.set_title('des nouveaux cas, décès et cas guérie du covide-19')
ax0.legend(['Confirmed', 'Deaths', 'Recovered'], loc='upper left', fontsize='x-large')
plt.xticks([cp_data_maroc.index[i] for i in range(0,len(cp_data_maroc.index),10)],rotation = 45)
plt.savefig('plot11.png')

ax0.plot()
```

```python
df_data_maroc_region = data_maroc_region
df_data_maroc_region.head()
```

```python
#creation nouvelle colon total
df_data_maroc_region = df_data_maroc_region.T
df_data_maroc_region['total'] = df_data_maroc_region.sum(axis = 1)
df_data_maroc_region.head()
```

```python
#creation de l'ensemble de donne
total = list(df_data_maroc_region['total'])
region = list(df_data_maroc_region.index)
d = {'region':region, 'latitude':latitude, 'longitude':longitude, 'total':total}
df = pd.DataFrame(data = d)
df
```

```python
#carte avec des marqeurs
import folium as fl
colors = ['gray', 'darkred', 'orange', 'beige', 'green', 'darkgreen',
          'lightgreen', 'darkblue', 'lightblue', 'purple', 'darkpurple', 'pink']
m = fl.Map(location = [31.172821, -7.336248], tiles = 'cartodbpositron', min_zoom = 1, zoom_start = 5)
for i in range(0, len(df)):
    folium.Marker(location = [df.iloc[i]['latitude'], df.iloc[i]['longitude']], icon=folium.Icon(color=colors[i]),
                 popup = '<li><bold> region: ' + str(df.iloc[i]['region']) +
                  '<li><bold> total: ' + str(df.iloc[i]['total'])).add_to(m)
m
```
